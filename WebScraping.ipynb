# WEB SCRAPING

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://resultados.as.com/resultados/futbol/primera/'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
```

image.png

```python
eq = soup.find_all('span', class_ = 'nombre-equipo')
equipos = list()
for i in eq:
  equipos.append(i.text)
print(len(equipos))
print(equipos)
```

```python
fch = soup.find_all('span', class_ = 'fecha')
fechas = list()
for i in fch:
  fechas.append(i.text)
print(len(fechas))
print(fechas)
```

```python
eq_1 = equipos[0::2] # obtenemos pares
eq_2 = equipos[1::2] # obtenemos impares
assert len(fechas) == len(eq_1) == len(eq_2)

df = pd.DataFrame({'Equipo 1': eq_1, 'Equipo 2': eq_2, 'Fecha': fechas}, list(range(1,11)))
df
```
